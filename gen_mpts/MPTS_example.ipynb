```python
import pandas as pd
import numpy as np
import os, os.path
import csv
from TICC_solver import TICC
import sys
import subprocess
from itertools import groupby
from scipy.stats import ttest_ind
import math
from scipy import stats
import random
```

```python
data = 'mimic' # dataset name
num_clusters = 6 # cluster number, generally selected by silhoutte albow, recommanded value range: integer in [5,20]
b = 0 # hyper-parameter for ticc

PATH = '../data/'+data+'/data_id_label.csv'

all_ins = pd.read_csv(PATH)
id_list = list(all_ins['visitID'].unique())

train_id_list = id_list[:97] + id_list[386:483] # training instances ids, can be changed according to train/test split 

all_ins_arr = all_ins[[col for col in all_ins.columns if 'X' in col]].to_numpy()

np.savetxt('../data/' +data+ '/data.txt', all_ins_arr, delimiter=',')
```

```python
## start TICC to generate cluster indices ##
fname = '../data/'+data+'/data.txt'
ticc = TICC(window_size=1, number_of_clusters=num_clusters, lambda_parameter=11e-2, beta=b, maxIters=100, threshold=2e-8,
            write_out_file=False, prefix_string="output_folder/", num_proc=1)
(cluster_assignment, cluster_MRFs) = ticc.fit(input_file=fname)

print(cluster_assignment)
np.savetxt('../result/ticc_'+data+'_c'+str(num_clusters)+'_b'+str(b)+'.txt', cluster_assignment, fmt='%d', delimiter=',')
## end TICC ## 
```

```python
# add ticc clusters to data

fname = '../result/ticc_'+data+'_c'+str(num_clusters)+'_b'+str(b)+'.txt'
cluster = []
with open(fname, 'r') as fd:
    reader = csv.reader(fd)
    for row in reader:
        cluster+=row
cluster = [int(i) for i in cluster]

all_ins['Cluster'] = cluster
```

```python
# generate cluster sequences of all instances

def createSeqMetric(df):
    SEQ_dict = {}
    
    num_group = len(df['Label'].unique())

    for g in range(num_group):
        globals()['seq_%s' % g] = []

    tot_seq_n = 0
    allseq_n = [0] * num_group

    for subj in set(df["visitID"]):

        temp_df = df.loc[df["visitID"] == subj]
        temp_seq = []

        subj_group = temp_df["Label"].values[0]

        # make temp sequence for subj
        for i in range(len(temp_df)): 
            temp_cat = f'{temp_df["Cluster"].iloc[i]}'
            temp_seq += [temp_cat]
        
        tot_seq_n += 1
        
        globals()['seq_%s' % subj_group].append(temp_seq)
        allseq_n[subj_group] += 1

        SEQ_dict[subj] = temp_seq
    
    # concate seq together
    allseq = []
    for idx in range(num_group):
        allseq.append(globals()['seq_%s' % idx])

    return (allseq, SEQ_dict, allseq_n)

seq = createSeqMetric(all_ins)
```

```python
# save sequences

for s in range(len(seq[0])):
    globals()['file_%s' % s] = open('../semi-result/'+data+'_seq_'+str(s)+'.txt', "w+")
    t = " "
    for i in range(len(seq[0][s])):
        temp = ''
        temp = t.join(seq[0][s][i])
        globals()['file_%s' % s].write('s' + str(i+1) + ' ' + temp + '\n')
    globals()['file_%s' % s].close()
```

```python
# generate cluster sequences of training instances, used for generating patterns

train_ins = all_ins.loc[all_ins['visitID'].isin(train_id_list)]
train_seq = createSeqMetric(train_ins)
for s in range(len(train_seq[0])):
    globals()['file_%s' % s] = open('../semi-result/'+data+'_train_seq_'+str(s)+'.txt', "w+")
    t = " "
    for i in range(len(train_seq[0][s])):
        temp = ''
        temp = t.join(train_seq[0][s][i])
        globals()['file_%s' % s].write('s' + str(i+1) + ' ' + temp + '\n')
    globals()['file_%s' % s].close()
```

```python
# make MPTS table for all instances
## use the patterns in pattern_all_list
## An example of pattern_all_list: [[4,4],[1,3,3,2],[2,8,8]]. [4,4] is a pattern with length 2

def makePatternTable(pattern_list):
    feature_num = len(pattern_list)
    feature_column = ['X'+ str(i) for i in range(feature_num)]

    all_ins_pat = []

    for ins in all_ins['visitID'].unique():
        ins_df = all_ins.loc[all_ins['visitID']==ins]
        ins_pat_df = pd.DataFrame(np.zeros((len(ins_df), feature_num)), columns=feature_column)
        ins_seq = ins_df.Cluster.values

        # get the last index of each pattern
        for f in range(feature_num):
            pattern = pattern_list[f]
            pat_len = len(pattern)
            for i in range(len(ins_seq)):
                if ins_seq[i] == pattern[0] and (i+pat_len-1)<=len(ins_seq):
                    if np.array_equal(ins_seq[i:(i+pat_len)], pattern):
                        ins_pat_df.iloc[i:(i+pat_len), f] = [1.0 for i in range(i,(i+pat_len))]

        all_ins_pat.append(ins_pat_df)
    all_ins_pat_df = pd.concat(all_ins_pat, axis=0, ignore_index=True)
    all_ins_pat_df['visitID'] = all_ins.visitID.values
    # all_ins_pat_df['time'] = all_ins.time.values
    all_ins_pat_df['Label'] = all_ins.Label.values
    all_ins_pat_df['Cluster'] = all_ins.Cluster.values
    return(all_ins_pat_df)

all_ins_pat_df = makePatternTable(pattern_all_list)

# save to csv
all_ins_pat_df.to_csv('../result/'+data+'_pattern_c'+str(num_clusters)+'.csv', index=False)
```
